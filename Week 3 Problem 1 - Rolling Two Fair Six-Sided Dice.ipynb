```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_trials = 5111
throws = np.random.randint(1, 7, size=(number_trials, 2), dtype=int)

realisations = throws.sum(axis=1)
# print('throws', throws)
# print('realisations', realisations)
```

```python
# Get the unique sums and their counts
unique_sums, counts = np.unique(realisations, return_counts=True)

# Normalize the counts
normalized_counts = counts / number_trials

# Create a bar plot of the normalized counts
plt.bar(unique_sums, normalized_counts)

# Set the title and labels
plt.title('Distribution of dice throws')
plt.xlabel('Sum')
plt.ylabel('Normalized frequency')

# Show the plot
plt.show()
```
